```python
import pandas as pd
import os
```

```python
pwd = os.getcwd()
print(pwd)
```

```python
# Load the data from the uploaded Excel file
data = pd.read_excel(pwd + "/data/selected_columns.xlsx")

# Display the first few rows of the data
data.head()
```

```python
# Correcting the "Single/Muplitple_Phase" column to merge 's' and 'S' into a single category 'Single'
data["Single/Muplitple_Phase"] = data["Single/Muplitple_Phase"].replace({'s': 'Single', 'S': 'Single', 'M': 'Multiple'})
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Plotting the data with corrected categories and hue based on "PhaseType"
plt.figure(figsize=(9, 6))
sns.boxplot(x=data["Single/Muplitple_Phase"], y=data["YieldStr(MPa)"], hue=data["PhaseType"])
plt.title("Yield Strength (MPa) vs Single/Multiple Phase")
plt.xlabel("Single/Multiple Phase")
plt.ylabel("Yield Strength (MPa)")
plt.legend(title="Phase Type")
plt.tight_layout()
#plt.savefig(pwd + "/YS_vs_Phase_type.png", dpi=1000)
plt.show()
```

```python
# Plotting the data with corrected categories and hue based on "PhaseType"
plt.figure(figsize=(9, 6))
sns.boxplot(x=data["Single/Muplitple_Phase"], y=data["YieldStr(MPa)"], hue=data["PhaseType"])
plt.title("Yield Strength (MPa) vs Single/Multiple Phase")
plt.xlabel("Single/Multiple Phase")
plt.ylabel("Yield Strength (MPa)")

# Position the legend at the 1/3 position on the x-axis and in the center on the y-axis
plt.legend(title="Phase Type", loc='center left', bbox_to_anchor=(1/4, 0.5))

plt.tight_layout()
plt.savefig(pwd + "/YS_vs_Phase_type.png", dpi=1000)
plt.show()
```
