```python
import itertools
import re

# Define a function to convert the string of elements into a list and generate binary combinations of two elements
def binary_combinations(elements_str):
    # Use regular expression to extract the elements from the input string
    elements = re.findall("'(\w+)'", elements_str)
    # Generate binary combinations of two elements
    return ['-'.join(sorted(pair)) for pair in itertools.combinations(elements, 2)]

# Apply the function to each row of the dataframe
df_comp['Binary_combinations'] = df_comp['Elements'].apply(binary_combinations)

df_comp.head()
```

```python
df = df_comp.copy()
```

```python
# convert each list element to a string by joining the elements with commas
df_comp['Binary_combinations'] = df_comp['Binary_combinations'].apply(lambda x: ','.join(x))

# use the str.split method to split the string into separate columns
df_comp = df_comp['Binary_combinations'].str.split(',', expand=True)

# rename the columns with binary_1, binary_2, etc.
df_comp.columns = ['binary_' + str(i+1) for i in range(len(df_comp.columns))]

df_comp.head()
```

```python
df_comp.shape
```

```python
df = pd.concat([df, df_comp], axis=1)
df.head()
```

```python
binary_columns = ['binary_1', 'binary_2', 'binary_3', 'binary_4', 'binary_5', 'binary_6', 'binary_7', 'binary_8', 'binary_9', 'binary_10']

# Create empty columns for the new values
for i in range(1, 11):
    df[f"volume_{i}"] = None
    df[f"density_{i}"] = None
    df[f"form_energy_pa_{i}"] = None

# Loop through each row in df2
for idx, row in df.iterrows():
    # Loop through each binary in the row
    for i, binary in enumerate(row[binary_columns], start=1):
        # Split the binary into two elements
        elements = binary.split('-')
        if len(elements) != 2:
            continue
        # Check for both possible combinations in df_binary
        matching_row = df_binary[df_binary['binary_systems'].isin([f"{elements[0]}-{elements[1]}", f"{elements[1]}-{elements[0]}"])]
        # If a matching row exists, add the values to the corresponding columns in df2
        if not matching_row.empty:
            df.at[idx, f"volume_{i}"] = matching_row.iloc[0]['volume']
            df.at[idx, f"density_{i}"] = matching_row.iloc[0]['density']
            df.at[idx, f"form_energy_pa_{i}"] = matching_row.iloc[0]['form_energy_pa']

df.head()
```

```python
# Get the last 30 column names
last_30_cols = df.columns[-30:]

# Convert the data type of the last 30 columns to numerical
df[last_30_cols] = df[last_30_cols].astype(float)
```

```python
# Create a function to sort the groups in a row
def sort_groups(row):
    groups = [
        {"binary": row[f"binary_{i}"],
         "volume": row[f"volume_{i}"],
         "density": row[f"density_{i}"],
         "form_energy_pa": row[f"form_energy_pa_{i}"]}
        for i in range(1, 11)
    ]

    sorted_groups = sorted(groups, key=lambda x: x["form_energy_pa"])

    for i, group in enumerate(sorted_groups, start=1):
        row[f"binary_{i}"] = group["binary"]
        row[f"volume_{i}"] = group["volume"]
        row[f"density_{i}"] = group["density"]
        row[f"form_energy_pa_{i}"] = group["form_energy_pa"]

    return row

# Apply the sorting function to each row
df = df.apply(sort_groups, axis=1)

df.head()
```

```python
df.to_excel(pwd + "/data/new_alloy_data.xlsx", index=False)
```
