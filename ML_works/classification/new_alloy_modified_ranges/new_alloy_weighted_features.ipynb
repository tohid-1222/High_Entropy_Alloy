```python
import numpy as np
import pandas as pd
import os
```

```python
pwd = os.getcwd()
print(pwd)
```

```python
data = pd.read_excel(pwd+"/data/new_alloy_data.xlsx")
print("Shape of the dataset is: ", data.shape)

data.head()
```

```python
df = data.copy()
```

```python
from pymatgen.core.composition import Composition

df['composition_dict'] = df['HEA_formula'].apply(lambda x: Composition(x).as_dict())

df.head()
```

```python
# Extract elements into separate columns
for idx, col_name in enumerate(['E1', 'E2', 'E3', 'E4', 'E5']):
    df[col_name] = df['composition_dict'].apply(lambda x: list(x.keys())[idx])

# Extract mole fractions into separate columns
for idx, col_name in enumerate(['C1', 'C2', 'C3', 'C4', 'C5']):
    df[col_name] = df['composition_dict'].apply(lambda x: list(x.values())[idx])

# Calculate sum of mole fractions
df = df.assign(sum_mole_fraction=df[['C1', 'C2', 'C3', 'C4', 'C5']].sum(axis=1))

df.head()
```

```python
# Create fraction columns
for idx, col_name in enumerate(['fraction_1', 'fraction_2', 'fraction_3', 'fraction_4', 'fraction_5']):
    df[col_name] = df[f'C{idx+1}'] / df['sum_mole_fraction']

df.head()
```

```python
# Create the 'elemental_dict' column
df['elemental_dict'] = df.apply(
    lambda row: {row['E1']: row['fraction_1'], row['E2']: row['fraction_2'],
                 row['E3']: row['fraction_3'], row['E4']: row['fraction_4'],
                 row['E5']: row['fraction_5']}, axis=1)

df.head()
```

```python
for idx in range(1, 11):
    binary_col = f'binary_{idx}'
    multiplier_col = f'multiplier_{idx}'
    
    df[multiplier_col] = df.apply(
        lambda row: row['elemental_dict'][row[binary_col].split('-')[0]] *
                    row['elemental_dict'][row[binary_col].split('-')[1]], axis=1)

df.head()
```

```python
for idx in range(1, 11):
    multiplier_col = f'multiplier_{idx}'
    volume_col = f'volume_{idx}'
    density_col = f'density_{idx}'
    form_energy_pa_col = f'form_energy_pa_{idx}'
    
    df[f'weighted_' + volume_col] = df[volume_col] * df[multiplier_col]
    df[f'weighted_' + density_col] = df[density_col] * df[multiplier_col]
    df[f'weighted_' + form_energy_pa_col] = df[form_energy_pa_col] * df[multiplier_col]

df.head()
```

```python
df.to_excel(pwd+"/data/new_alloy_weighted_features.xlsx", index=False)
```
